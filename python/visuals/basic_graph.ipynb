```python
# Making a basic Bokeh graph
# importing bokeh

from bokeh.io import output_file, show
from bokeh.layouts import column
from bokeh.plotting import figure


# prepare some data
x=[1,2,3,4,5]
y=[6,7,8,9,10]

# prepare the output file
output_file("Line.html")

# create figure object
f=figure()
g=figure()
h=figure()

# figure g and h points
xgh=[3,7.5,10]
ygh=[3,6,10]

# create line plot
f.line(x,y)

g.circle(xgh,ygh)

h.triangle(xgh,ygh)

show(column(f,g,h))
```
