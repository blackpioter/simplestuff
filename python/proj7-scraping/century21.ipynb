```python
import requests
from bs4 import BeautifulSoup
import pandas

r = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})

c = r.content

soup = BeautifulSoup(c, "html.parser")

all=soup.findAll("div",{"class":"propertyRow"})

all[0].find("h4",{"class":"propPrice"}).text.strip()
```

```python
all[0]
```

```python
l=[]
for item in all:
    d={}
    d["Price"]=item.find("h4",{"class":"propPrice"}).text.strip()
    addr=item.findAll("span",{"class":"propAddressCollapse"})
    d["Address"]=addr[0].text
    d["Location"]=addr[1].text
    try:
        d["Beds"]=item.find("span",{"class":"infoBed"}).find("b").text
    except:
        d["Beds"]=None
    try:
        d["Area"]=item.find("span",{"class":"infoSqFt"}).find("b").text
    except:
        d["Area"]=None        
    try:
        d["Full Baths"]=item.find("span",{"class":"infoValueFullBath"}).find("b").text
    except:
        d["Full Baths"]=None
    try:
        d["Half Baths"]=item.find("span",{"class":"infoValueHalfBath"}).find("b").text
    except:
        d["Half Baths"]=None
        
    try:
        for column_group in item.findAll("div",{"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.findAll("span", {"class":"featureGroup"}),column_group.findAll("span",{"class":"featureName"})):
                if "Lot Size" in feature_group.text:
                    d["Lot Size"]=feature_name.text
    except:
        d["Lot Size"]=None
    
    l.append(d)
```

```python
l
```

```python
df=pandas.DataFrame(l)
df
```

```python
df.to_csv("Century")
```
